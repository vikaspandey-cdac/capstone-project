```python
data1['release_date'] = pd.to_datetime(data1['release_date'], errors='coerce')
```

```python
data1['year'] = data1['release_date'].dt.year
```

```python
data1['year'].value_counts().sort_index()
```

```python
# Getting only 2017 movies as we already have movies up to the year 2016 in preprocessing 1 file. 
# We don't have enough data for the movies from 2018, 2019 and 2020. 
# We'll deal with it in the upcoming preprocessing files
up_data1 = data1.loc[data1.year == 2017,['genres','id','title','year']] # up_data1 --> updated data1
```

```python
up_data1
```

```python
up_data1['id'] = up_data1['id'].astype(int)
up_data = pd.merge(up_data1, data2, on='id')
```

```python
pd.set_option('display.max_colwidth', 75)
up_data
```

```python
# evaluates an expression node or a string containing a Python literal or container display
import ast
up_data['genres'] = up_data['genres'].map(lambda x: ast.literal_eval(x))
up_data['cast'] = up_data['cast'].map(lambda x: ast.literal_eval(x))
up_data['crew'] = up_data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))
```

```python
up_data['genres_list'] = up_data['genres'].map(lambda x: make_genresList(x))
```

```python
up_data['genres_list'].head()
```

```python
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return (casts[0])
```

```python
up_data['actor_1_name'] = up_data['cast'].map(lambda x: get_actor1(x))
```

```python
up_data['actor_1_name'].head()
```

```python
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
```

```python
up_data['actor_2_name'] = up_data['cast'].map(lambda x: get_actor2(x))
```

```python
up_data['actor_2_name'].head()
```

```python
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
up_data['actor_3_name'] = up_data['cast'].map(lambda x: get_actor3(x))
```

```python
up_data['actor_3_name'].head()
```

```python
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
up_data['director_name'] = up_data['crew'].map(lambda x: get_directors(x))
```

```python
up_data['director_name'].head()
```

```python
movie = up_data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres_list','title']]
```

```python
movie.head()
```

```python
movie.isna().sum()
```

```python
movie = movie.dropna(how='any')
```

```python
movie.isna().sum()
```

```python
movie = movie.rename(columns={'genres_list':'genres'})
movie = movie.rename(columns={'title':'movie_title'})
```

```python
movie['movie_title'] = movie['movie_title'].str.lower()
```

```python
movie['comb'] = movie['actor_1_name'] + ' ' + movie['actor_2_name'] + ' '+ movie['actor_3_name'] + ' '+ movie['director_name'] +' ' + movie['genres']
```
